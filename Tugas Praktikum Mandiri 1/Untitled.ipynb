```python
import pandas as pd

# Membaca data dari file CSV
data = pd.read_csv("Data_BMI.csv")

# Hitung statistik berdasarkan Gender untuk kolom berat (Weight)
statistik_berat = data.groupby('Gender')['Weight'].agg(['count', 'mean', 'max', 'min'])
statistik_berat.columns = ['Jumlah Data', 'Rata-rata Berat', 'Berat Maksimum', 'Berat Minimum']

# Hitung statistik berdasarkan Gender untuk kolom tinggi (Height)
statistik_tinggi = data.groupby('Gender')['Height'].agg(['mean'])
statistik_tinggi.columns = ['Rata-rata Tinggi']

# Gabungkan kedua statistik menjadi satu DataFrame
statistik_gabungan = pd.concat([statistik_berat, statistik_tinggi], axis=1)

# Tampilkan statistik
print("\nStatistik Berdasarkan Gender:")
print(statistik_gabungan)
```

```python
# Tampilkan grafik Index BMI berdasarkan gender

import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("Data_BMI.csv")
data['Gender'] = data['Gender'].map({'Female': 0, 'Male': 1})
data['BMI'] = data['Weight'] / ((data['Height'] / 100) ** 2)
plt.figure(figsize=(8, 6))
plt.scatter(data['Gender'], data['BMI'], c=data['Index'], cmap='viridis')
plt.colorbar(label='Index BMI')
plt.xticks([0, 1], ['Female', 'Male'])
plt.xlabel('Gender')
plt.ylabel('Index BMI')
plt.title('Index BMI Berdasarkan Gender')
plt.show()
```
